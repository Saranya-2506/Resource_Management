```python
import numpy as np

for col in numerical_col:
    Q1=merge_df[col].quantile(0.25)
    Q3=merge_df[col].quantile(0.75)
    IQR=Q3-Q1

    lower_bound=Q1-(1.5)*IQR
    upper_bound=Q3+(1.5)*IQR

    # CAP VALUES IN UPPER AND LOWER BOUND VALUES

    merge_df[col]=np.where(merge_df[col]<lower_bound,lower_bound,merge_df[col])
    merge_df[col]=np.where(merge_df[col]>upper_bound,upper_bound,merge_df[col])
```

```python
# LowerCase all Text
import re

def preprocess_text(text):

    text=text.lower()

# Remove pantuation and numbers

    text=re.sub(r'[^a-z\s]','',text)

#Tokenize

    tokens=word_tokenize(text)

#Remove stop words and lemmatize

    stop_words=set(stopwords.words('english'))

    lemmatizer=WordNetLemmatizer()

    cleaned_tokens=[lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

#Join tokens back into single string for storage  or further process

    return(' '.join(cleaned_tokens))

```

```python
# Apply function to the column


merge_df['cleaned_text']=merge_df["NIC Name"].apply(preprocess_text)
print(merge_df[['NIC Name','cleaned_text']].head())
```

```python
# Term Frequency-Inverse Document Frequency to Vectorize the text

# Feature Extraction
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Instantiate the vectorizer()

vectorizer=TfidfVectorizer()
#Fit and transorm cleaned text data
tfid_matrix=vectorizer.fit_transform(merge_df['cleaned_text'])

# view the result

feature_names=vectorizer.get_feature_names_out()

tdidf_df=pd.DataFrame(tfid_matrix.toarray(),columns=feature_names)
tdidf_df

```

```python
# TF-IDF Vectorization 
import pickle

with open("tfidf.pkl","wb") as f:
    pickle.dump(vectorizer,f)
```

```python
#Elbow Technique
import numpy as np
import matplotlib.pyplot as pt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,davies_bouldin_score

x=tdidf_df

# Calculate inertia/WCSS Within -Cluster sum of squares

wcss=[]
k_range=range(5,15)


for k in k_range:
    kmeans=KMeans(n_clusters=k,random_state=42,n_init=10)
# Fit the model to the data
    kmeans.fit(x)
    y_pred=kmeans.labels_
    #Append inertia to the list
    t=kmeans.inertia_
    wcss.append(t)
    sil_score=silhouette_score(x,y_pred)
    davies_score=davies_bouldin_score(x,y_pred)
    print(f"silhouette score:{sil_score:.2f}")
    print(f"Davies bouldin score:{davies_score:.2f}")
#plot the wcss values againstnk values
pt.figure(figsize=(10,8))

pt.plot(k_range,wcss,marker='o',linestyle='-')
pt.title("The Elbow method using Inertia ")
pt.xlabel("Number of clusters(k)")
pt.ylabel("Inertia(WCSS)")
pt.grid(True)
pt.show()
```

```python
# K means cluster

import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as pt


X=tdidf_df
kmeans=KMeans(n_clusters=8,random_state=42,n_init=10)
clusters=kmeans.fit_predict(X)

merge_df['Cluster']=clusters
merge_df




```
